# **Linear Regression**

The idea with linear regression is to fit a line according to set of data. This could help us to describe some trends related with our information, thus we could predict some unobserverd values.

Linear regression use the notion of **least squares**, where it minimized the squared error between the line and each point, this occurs having in mind the sum of this least squares

**See maximum likehood estimation**

## **Error**

We can measure the error with r-squared (coefficient of determination), with this we can see how our line follow well the variation. We can compute r-squared as follows:
$$\begin{align}
R^2&= 1 - \frac{\mathrm{sum of squared regression (SSR)}}{\mathrm{total sum of square (SST)}}\\
&=1-\frac{\sum({y_i}-\hat{y_i})^2}{\sum(y_i-\bar{y})^2},
\end{align}$$
where the sum squared regression is the sum of the residuals squared (sum of the difference between the original data and data in line all squared), and the total sum of squares (is the sum of the difference between the original data and the mean of the data all squared). Therefore, we have that $$0\leq R^2 \leq 1.$$

Near to zero the linear regression is poor.
Near to one the linear regression is good.

## **Example**

We are going to create a random distribution. At first we are going to take random values to describe one of the axis, an then we create a function in order to describe a trend as a line.

So we are going to create a dataset with the amount of days lonely by year according to the age

```python
import numpy as np
from pylab import *
from scipy import stats #To call the linear regression
import matplotlib.pyplot as plt #To plot
```

```python
age = np.random.normal(33.0, 9.0, 200)
hoursLonely = 100 + (age + np.random.normal(0, 0.1, 200)) * 2

scatter(age, hoursLonely)
```

Now we are going to use stats in order to call the linear regression

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(age, hoursLonely)
```

After that, we can see the R^2 is near to 1, thus our line fits really well with the data:

```python
R2=r_value ** 2
R2
```

Let us plot the line and the data given before:

```python
#This function describe our equation of the linear regression
def line_pred(x):
  return slope * x + intercept
```

```python
#Plotting the linear regression
linearReg = line_pred(age)
plt.scatter(age, hoursLonely)
plt.plot(age, linearReg, color='r')
plt.show()
```

## **With different dispersion**

Let us compare several linear regression differente dispersion in the data:

```python
for i in range(4):
  n=0.1+1.5*i
  #Data
  age = np.random.normal(33.0, 9.0, 200)
  hoursLonely = 100 + (age + np.random.normal(0, n, 200)) * 2
  scatter(age, hoursLonely)
  #Linear Regression
  slope, intercept, r_value, p_value, std_err = stats.linregress(age, hoursLonely)
  #R^2
  R2 = r_value ** 2
  #Plotting the linear regression
  linearReg = line_pred(age)
  plt.title('Dispersion: '+ str(n)+'\nR^2= ' + str(R2))
  plt.scatter(age, hoursLonely)
  plt.plot(age, linearReg, color='r')
  plt.show()
```

We can see, if the data is more disperse, then the $R^2$ dicreases.
